# Bisection Method 

```python
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def func(x):
  return x**6 -x -1
```

```python
a = 1.0000
b = 2.0000

table = pd.DataFrame(columns=['a', 'b', 'c = (a + b)/2', 'f(c)', 'b-c = c-a', 'f(a)*f(c)'])
for i in range(15):
    c = float( (a + b)/2 )
    f_a = func(a)
    f_c = func(c)
    acc = b-c
    if((f_a)*(f_c) < 0):
        flag = 'Negative'
        b = c
    else:
        flag = 'Positive'
        a = c
    table.loc[i] = [a,b,c,f_c,acc,flag]

plt.stem([i for i in range(15)], table['c = (a + b)/2'])
plt.title("Convergence of root between 1 and 2")
plt.show()

table
```

```python
a = -1.0000
b =  0.0000

table = pd.DataFrame(columns=['a', 'b', 'c = (a + b)/2', 'f(c)', 'b-c = c-a', 'f(a)*f(c)'])
for i in range(15):
    c = float( (a + b)/2 )
    f_a = a**6 -a -1
    f_c = c**6 -c -1
    acc = b-c
    if((f_a)*(f_c) < 0):
        flag = 'Negative'
        b = c
    else:
        flag = 'Positive'
        a = c
    table.loc[i] = [a,b,c,f_c,acc,flag]
    
plt.stem([i for i in range(15)], table['c = (a + b)/2'])
plt.title("Convergence of root between -1 and 0")
plt.show()

table  
```
