# Newton-Raphson Method - 201601416 & 201601410

```python
import numpy  as np
import pandas as pd
import math
import matplotlib.pyplot as plt
```

```python
table = pd.DataFrame(columns=['n', 'x(n)', 'f(x)', "f'(x)", 'x(n) - x(n-1)', 'x(n+1)', 'x(n+1) - x(n)'])

n    = 0
x_n  = 1.5
f_x  = x_n**6 -x_n -1
dfdx = 6*(x_n**5) - 1
dif1 = 0
x_n1 = x_n - (f_x/dfdx)
dif2 = x_n1 - x_n
table.loc[0] = [n, x_n, f_x, dfdx, dif1, x_n1, dif2]

while(1):
  n += 1
  dif1 = x_n1 - x_n
  x_n = x_n1
  f_x  = x_n**6 -x_n -1
  dfdx = 6*(x_n**5) - 1
  x_n1 = x_n - (f_x/dfdx)
  if2 = x_n1 - x_n
  table.loc[n] = [n, x_n, f_x, dfdx, dif1, x_n1, dif2]
  if(abs(dif1) < 0.0001):
    break

plt.stem(table['n'], table['x(n+1)'])
plt.title("Convergence of root between 1 and 2")
plt.show()
plt.plot(table["n"], table["f'(x)"])
plt.title("Graph of f'(x)")
plt.show()
table
```
